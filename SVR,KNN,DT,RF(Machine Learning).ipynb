# SVR,KNN,DT,RF(Regression Algorithms)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv(r"D:\FSDS\april(2025)\30th-1.SVR, KNN\Polynomial regression\1.POLYNOMIAL REGRESSION\emp_sal.csv")
dataset
```

```python
x=dataset.iloc[:,1:2].values 
y=dataset.iloc[:,2].values
```

```python
x
```

```python
y
```

# Support Vector Regression

```python
# svm model
from sklearn.svm import SVR
svr_regressor = SVR(kernel='poly',degree = 5)
svr_regressor.fit(x,y)

#SVR regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,svr_regressor.predict(x),color='blue')
plt.title('SVR Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

svr_model_pred = svr_regressor.predict([[6.5]])
print(svr_model_pred)
```

## SVR WITH DEGREE = 5

```python
# svm model
from sklearn.svm import SVR
svr_regressor = SVR(kernel='poly',degree = 5,gamma = 'scale' )
svr_regressor.fit(x,y)
```

```python
#SVR regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,svr_regressor.predict(x),color='blue')
plt.title('SVR Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
svr_model_pred = svr_regressor.predict([[6.5]])
print(svr_model_pred)
```

## SVR With Degree = 4 and gamma = "auto" (better accuracy) than degree = 5 

```python
# svm model
from sklearn.svm import SVR
svr_regressor = SVR(kernel='poly',degree = 4,gamma = 'auto' )
svr_regressor.fit(x,y)

#SVR regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,svr_regressor.predict(x),color='blue')
plt.title('SVR Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

svr_model_pred = svr_regressor.predict([[6.5]])
print(svr_model_pred)
```

# KNearest-NEIGHBORS

```python
# knn model 
from sklearn.neighbors import KNeighborsRegressor
knn_reg_model = KNeighborsRegressor()
knn_reg_model.fit(x,y)
```

```python
#KNN regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,knn_reg_model.predict(x),color='blue')
plt.title('KNN Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

knn_reg_pred = knn_reg_model.predict([[6.5]])
print(knn_reg_pred)
```

```python
# knn model 
from sklearn.neighbors import KNeighborsRegressor
knn_reg_model = KNeighborsRegressor(n_neighbors=6,p=2)
knn_reg_model.fit(x,y)

#KNN regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,knn_reg_model.predict(x),color='blue')
plt.title('KNN Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

knn_reg_pred = knn_reg_model.predict([[6.5]])
print(knn_reg_pred)
```

```python
# knn model 
from sklearn.neighbors import KNeighborsRegressor
knn_reg_model = KNeighborsRegressor(n_neighbors=4,weights='distance')
knn_reg_model.fit(x,y)

#KNN regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,knn_reg_model.predict(x),color='blue')
plt.title('KNN Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

knn_reg_pred = knn_reg_model.predict([[6.5]])
print(knn_reg_pred)
```

```python
# knn model 
from sklearn.neighbors import KNeighborsRegressor
knn_reg_model = KNeighborsRegressor(n_neighbors=5, weights='distance', p=2)
knn_reg_model.fit(x,y)
```

```python
#KNN regression visualization
plt.scatter(x,y,color='red')
plt.plot(x,knn_reg_model.predict(x),color='blue')
plt.title('KNN Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

knn_reg_pred = knn_reg_model.predict([[6.5]])
print(knn_reg_pred)
```
